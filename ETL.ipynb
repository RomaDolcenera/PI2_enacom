```python
#visualizo el Dataframe y encuentro '*' en Año y Trimestre. Reemplazo el *
access_tec = accesos_tec.copy()
access_tec['Año'].unique() 
access_tec['Año'] = access_tec['Año'].str.replace('[^0-9]', '').str.rstrip('*')
access_tec['Año'] = access_tec['Año'].astype(int)
#access_tec.head(5)
access_tec['Año'].unique() 
```
