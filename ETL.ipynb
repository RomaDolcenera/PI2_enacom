```python
#visualizo el Dataframe y encuentro '*' en Año y Trimestre. Reemplazo el *
accesos_tec = accesos_tec.copy()
accesos_tec['Año'].unique() 
accesos_tec['Año'] = accesos_tec['Año'].str.replace('[^0-9]', '').str.rstrip('*')
accesos_tec['Año'] = accesos_tec['Año'].astype(int)
#access_tec.head(5)
accesos_tec['Año'].unique() 
```

```python
accesos_tec.info()
```

```python
accesos_tec['Trimestre'] = accesos_tec['Trimestre'].str.replace('[^0-9]', '').str.rstrip('*')
accesos_tec['Trimestre'] = accesos_tec['Trimestre'].astype(int)
```

```python
accesos_tec.info()
```

```python
# Combino el df previo creado con las mismas columnas
df2_comb = pd.merge(accesos_tec, df1_comb, on=['Año', 'Trimestre', 'Provincia'])
df2_comb.info()
```

```python
df2_comb.sample(5)
```

```python
# Paso todos los datos a interger
df2_comb = df2_comb.astype({
    'Provincia': 'string',
    'ADSL': 'int',
    'Cablemodem': 'int',
    'Wireless': 'int',
    'Otros': 'int',
    'Total': 'int',
    'Fibra óptica': 'int',
    'Accesos por cada 100 hab': 'int',
    'Accesos por cada 100 hogares': 'int'
})

```

```python
df2_comb.info()
```

```python
#Unifico DF con el mismo tipo de informacion.
df3_comb = pd.merge(df2_comb, accesos_baf_pcia, on=['Año', 'Trimestre', 'Provincia'])
df3_comb.info()
```

```python
# Evaluo los resultados del trabajo
igual_1 = (df4_comb['Total'] == df4_comb['Total_x']).sum()
igual_2 = (df4_comb['Total'] == df4_comb['Total_y']).sum()
igual_3 = (df4_comb['Total_x'] == df4_comb['Total_y']).sum()

print(igual_1)
print(igual_2)
print(igual_3)
```

```python
#Elimino los '.'
df4_nac['Hasta 512 kbps'] = df4_nac['Hasta 512 kbps'].astype(str).str.replace('.',',')
```

```python
# Creo una Funcion para reemplazar los '+' y los '-'
def renombre_col(columna):
    columna = columna.replace('+', 'Entre')
    columna = columna.replace('-','y')
    return columna
```
